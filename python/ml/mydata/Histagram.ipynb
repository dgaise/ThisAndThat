```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline 
```

```python
my_df = pd.DataFrame(randn(500, 4), columns=["Mon","Tue","Wed","Thur"])
my_df
```

```python
my_df.hist(bins=10,figsize=(6,8), legend=True)
```

```python
my_df[["Tue","Wed"]].hist(bins=50,legend=True, grid=False)
```

```python
my_df.Tue.plot(kind="hist") # No bars
```

```python
my_df.abs().plot(kind="area") # area
```

```python
my_df.abs().plot(kind="area", alpha=0.4, title="My plot") # less bright, with title 
```

```python
my_df[["Mon","Thur"]].abs().plot(kind="area", 
                                 alpha=0.4,   # less bright, with title 
                                 title="My plot", 
                                 grid=True) 
```

```python
my_df.abs().plot(kind="bar", 
                                 alpha=0.7,   # less bright, with title 
                                 title="My plot", 
                                 grid=False,
                                 stacked=True)                                                                    
```

```python
my_df[["Mon","Wed"]].abs().plot(kind="line", alpha=0.9,   # less bright, with title 
        title="My Line plot", 
        grid=True ,
        lw=0.5 , ##line width
        figsize=(12,4)
                               )
```

```python
my_df.plot(kind="scatter", x="Mon", y="Wed",
                 c="Tue",#cmap="magma", ### colormaps - https://matplotlib.org/stable/users/explain/colors/colormaps.html
                         cmap="spring", #Greens, plasma, viridis
        alpha=0.9,   # less bright, with title 
        title="My Line plot", 
        grid=True , 
        lw=0.5 , ##line width
        figsize=(12,4)
                               )
```

```python
my_df.abs().plot(kind="scatter", x="Mon", y="Tue", s=my_df["Wed"]* 200,
        alpha=0.4,   
        title="My Line plot", 
        grid=True , 
        figsize=(12,4)
                               )
```

```python
my_df.plot(kind="box" ,  cmap="spring"  ,
        title="My Box plot", 
        grid=True , 
        figsize=(12,4)                        )
```

```python
my_df.plot(kind="hexbin" , x="Mon", y="Wed", gridsize=10,
        title="My Box plot", 
        grid=True , 
        figsize=(12,4)                        )
```

```python
my_df.plot(kind="kde", title="My KDE plot", 
        grid=True , 
        figsize=(12,4)                        )
```

```python
my_df.Mon.plot.density(title="My KDE plot", 
        grid=True , 
        figsize=(12,4)  )
```
